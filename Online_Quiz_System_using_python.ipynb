```python
# Online Quiz System Using Python
class Quiz:
    def __init__(self, title, questions):
        self.title = title
        self.questions = questions
        self.score = 0

    def display_question(self, question_number):
        print(f"\nQuestion {question_number + 1}: {self.questions[question_number]['question']}")
        for i, option in enumerate(self.questions[question_number]['options']):
            print(f"{chr(65 + i)}. {option}")

    def take_quiz(self):
        print(f"Welcome to the {self.title} quiz!")
        for i in range(len(self.questions)):
            self.display_question(i)
            answer = input("Enter your answer (A, B, C, or D): ").upper()
            if answer == self.questions[i]['answer']:
                print("Correct!")
                self.score += 1
            else:
                print("Incorrect!")
        print(f"\nQuiz completed. Your score: {self.score}/{len(self.questions)}")

class User:
    def __init__(self, username):
        self.username = username
        self.quiz_scores = {}

    def take_quiz(self, quiz):
        print(f"\nWelcome, {self.username}!")
        quiz.take_quiz()
        self.quiz_scores[quiz.title] = quiz.score

if __name__ == "__main__":
    quiz_title = "General Knowledge"
    quiz_questions = [
        {
            "question": "What is the capital of France?",
            "options": ["A. Paris", "B. London", "C. Berlin", "D. Rome"],
            "answer": "A"
        },
        {
            "question": "What is the largest planet in our solar system?",
            "options": ["A. Jupiter", "B. Saturn", "C. Neptune", "D. Earth"],
            "answer": "A"
        },
        {
            "question": "Which country is known as the Land of the Rising Sun?",
            "options": ["A. China", "B. Japan", "C. South Korea", "D. Thailand"],
            "answer": "B"
        },
        {
            "question": "What is the chemical symbol for water?",
            "options": ["A. H2O", "B. CO2", "C. O2", "D. HCl"],
            "answer": "A"
        },
        {
            "question": "Who painted the Mona Lisa?",
            "options": ["A. Leonardo da Vinci", "B. Michelangelo", "C. Pablo Picasso", "D. Vincent van Gogh"],
            "answer": "A"
        }
    ]

    quiz = Quiz(quiz_title, quiz_questions)

    username = input("Enter your name: ")
    user = User(username)
    user.take_quiz(quiz)
```
